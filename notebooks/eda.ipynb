# 🏠 House Price EDA
Exploratory Data Analysis for the House Price Stacking Regressor.

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

df = pd.read_csv('../data/houses.csv')
print(df.shape)
df.head()
```

## 1. Target Distribution

```python
px.histogram(df, x='Sale_Price', nbins=60, title='Sale Price Distribution', template='plotly_white').show()
```

## 2. Price by Neighborhood

```python
px.box(df, x='Neighborhood', y='Sale_Price', color='Neighborhood', title='Sale Price by Neighborhood', template='plotly_white').show()
```

## 3. Square Footage vs Price

```python
px.scatter(df, x='Total_SqFt', y='Sale_Price', color='Neighborhood', title='Total SqFt vs Sale Price', template='plotly_white').show()
```

## 4. Overall Quality vs Median Price

```python
avg = df.groupby('Overall_Qual')['Sale_Price'].median().reset_index()
px.bar(avg, x='Overall_Qual', y='Sale_Price', color='Overall_Qual', title='Median Price by Overall Quality', template='plotly_white').show()
```

## 5. Correlation Heatmap

```python
num = df.select_dtypes(include=[float, int])
corr = num.corr().round(2)
fig = go.Figure(go.Heatmap(z=corr.values, x=corr.columns.tolist(), y=corr.index.tolist(), colorscale='RdBu', zmid=0, text=corr.values, texttemplate='%{text}'))
fig.update_layout(title='Feature Correlation Heatmap', template='plotly_white', height=550)
fig.show()
```
